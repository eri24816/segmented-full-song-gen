```python
from pathlib import Path

from music_data_analysis import Pianoroll
from segment_full_song.data.dataset import FullSongPianorollDataset
from segment_full_song.models.token_sequence import TokenSequence
import random
from tqdm import tqdm

ds = FullSongPianorollDataset(Path("../../pop80k_k"), split='train')

sum_bars = 0
sum_tokens = 0
for i in tqdm(range(1000)):
    i = random.randint(0, len(ds)-1)
    pr = Pianoroll.from_midi(ds[i].read_midi('synced_midi'))
    sum_bars += pr.duration // 32
    ts = TokenSequence.from_pianorolls([pr], 64)
    sum_tokens += ts.length

print(sum_bars / 1000, sum_tokens / 1000)
```
